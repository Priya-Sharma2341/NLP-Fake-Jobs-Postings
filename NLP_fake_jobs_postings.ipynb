```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#Vectorization
cv=TfidfVectorizer(max_features=1000)
w=cv.fit_transform(data["text"])
df=pd.DataFrame(w.toarray(),columns=cv.get_feature_names())
data.drop(["text"],axis=1,inplace=True)
main_data=pd.concat([df,data],axis=1)
```

```python
main_data.head()
```

```python
x1=main_data.iloc[:,:-1]
y1=main_data.iloc[:,-1]

#Splitting data in test and train
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.20,random_state=160111)
print(x1_train.shape)
print(y1_train.shape)
print(x1_test.shape)
print(y1_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_jobs=3,oob_score=True,n_estimators=100,criterion="entropy")
model=rfc.fit(x1_train,y1_train)
```

```python
x1
```

```python
from sklearn.metrics import accuracy_score
```

```python
pred=rfc.predict(x1_test)
score=accuracy_score(y1_test,pred)
score
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print("Classification Report\n")
print(classification_report(y1_test,pred))
print("Confusion Matrix\n")
print(confusion_matrix(y1_test,pred))
```
